```python
import struct
import numpy as np

def load_images(filepath):
    with open(filepath, 'rb') as f:
        magic, num, rows, cols = struct.unpack('>IIII', f.read(16))
        return np.frombuffer(f.read(), dtype=np.uint8).reshape(num, rows, cols)

def load_labels(filepath):
    with open(filepath, 'rb') as f:
        magic, num = struct.unpack('>II', f.read(8))
        return np.frombuffer(f.read(), dtype=np.uint8)

# Charger les 4 fichiers
x_train = load_images('gzip\emnist-byclass-train-images-idx3-ubyte\emnist-byclass-train-images-idx3-ubyte')
y_train = load_labels('gzip\emnist-byclass-train-labels-idx1-ubyte\emnist-byclass-train-labels-idx1-ubyte')

x_test  = load_images('gzip\emnist-byclass-test-images-idx3-ubyte\emnist-byclass-test-images-idx3-ubyte')
y_test  = load_labels('gzip\emnist-byclass-test-labels-idx1-ubyte\emnist-byclass-test-labels-idx1-ubyte')

print(f"Train : {x_train.shape}  →  {len(np.unique(y_train))} classes")
print(f"Test  : {x_test.shape}")
```
